```python
# Your code goes here
import pandas as pd

url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)

df1.head()
df1.shape
```

```python
df2.head()
```

```python
df3.head()
```

```python
df1.columns = df1.columns.str.replace("ST", "state")

df1.columns = (
    df1.columns
      .str.lower()
      .str.replace(" ", "_")
)

df1.columns

df2.columns = df2.columns.str.replace("ST", "state")

df2.columns = (
    df2.columns
      .str.lower()
      .str.replace(" ", "_")
)

df2.columns

df3.columns = df3.columns.str.replace("ST", "state")

df3.columns = (
    df3.columns
      .str.lower()
      .str.replace(" ", "_")
)

df3.columns
```

```python
df_all = pd.concat([df1, df2, df3], axis=0)
```

```python
print(df_all.shape)
df_all.head(20)
```

```python
gender_map = {
    "M": "M",
    "Male": "M",
    "male": "M",
    "F": "F",
    "Femal": "F",
    "Female": "F",
    "female": "F"
}

# Standardize gender column using map()
df_all['gender'] = df_all['gender'].replace(gender_map)

state_map = {
    "AZ": "Arizona",
    "Cali": "California",
    "WA": "Washington",
}

# Standardize gender column using map()
df_all['state'] = df_all['state'].replace(state_map)

df_all['education'] = df_all['education'].replace("Bachelors", "Bachelor")

# Remove '%' character from customer_lifetime_value
df_all['customer_lifetime_value'] = df_all['customer_lifetime_value'].str.replace('%', '', regex=False)


vehicle_map = {
    "Sports Car": "Luxury",
    "Luxury SUV": "Luxury",
    "Luxury Car": "Luxury",
}

# Standardize gender column using map()
df_all['vehicle_class'] = df_all['vehicle_class'].replace(vehicle_map)

df_all.head(20)
```

```python
df_all.dtypes
```

```python
df_all["customer_lifetime_value"] = df_all["customer_lifetime_value"].astype(float)


def remove_bar(entry):
    # Take the middle value after splitting by "/"
    #print(entry)
    if pd.isnull(entry):
        return entry
    elif type(entry) is int:
        return float(entry)
    else:
        return float(entry.split("/")[1])

df_all["number_of_open_complaints"] = df_all["number_of_open_complaints"].apply(remove_bar)
#df_all["number_of_open_complaints"] = df_all["number_of_open_complaints"].astype(float)

df_all["number_of_open_complaints"]
```

```python
df_all.isnull().sum()
```

```python
df_all = df_all.dropna(how="all")

df_all.isnull().sum()
```

```python
df_all.shape
```

```python
df_all = df_all.drop("customer_lifetime_value", axis=1) #dropped this because too many nan values

```

```python
df_all.shape
df_all.isnull().sum()
```

```python
df_all["gender"].value_counts()
df_all["gender"] = df_all["gender"].fillna("F")
df_all.isnull().sum()
```

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*
